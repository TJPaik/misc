```python
import numpy as np
import matplotlib.pyplot as plt


n_points = [2, 5, 7, 10, 15, 20]
```

```python
spheres = [np.load(f'/mount/paiktj/sphere{i}.npy') for i in [2,3,4,5]]
```

```python
spheres[0].shape
```

```python
fit, axs = plt.subplots(6, 4, figsize=(10, 10))
# fit, axs = plt.subplots(4, 6)
for i in range(4):
    for j in range(6):
        axs[j][i].plot(spheres[i][j].mean(axis=0))
        axs[j][i].set(xlabel=f"S^{i+1}", ylabel=f"n_points: {n_points[j]}")
        axs[j][i].axhline(-1, color='r', lw=0.5)
        axs[j][i].axhline(0, color='r', lw=0.5)
        axs[j][i].axhline(1, color='r', lw=0.5)
        axs[j][i].axhline(2, color='r', lw=0.5)

for ax in axs.flat:
    ax.label_outer()
plt.suptitle('Euler mean')
plt.tight_layout()
plt.show()
```

```python
fit, axs = plt.subplots(6, 4, figsize=(10, 10))
# fit, axs = plt.subplots(4, 6)
for i in range(4):
    for j in range(6):
        axs[j][i].plot(spheres[i][j].std(axis=0))
        axs[j][i].set(xlabel=f"S^{i+1}", ylabel=f"n_points: {n_points[j]}")
        axs[j][i].axhline(-1, color='r', lw=0.5)
        axs[j][i].axhline(0, color='r', lw=0.5)
        axs[j][i].axhline(1, color='r', lw=0.5)
        axs[j][i].axhline(2, color='r', lw=0.5)

for ax in axs.flat:
    ax.label_outer()
plt.suptitle('Euler std')
plt.tight_layout()
plt.show()
```

```python
n_points = [4,8,16,32, 64, 128, 256, 512, 1024]
S1_bettis = [np.load(f'/mount/paiktj/first_betti_sphere_S1_{i}.npy') for i in n_points]
```

```python
S1_bettis[-1].shape
```

```python
fig, axs = plt.subplots(2, len(n_points), figsize=(20, 5))
for i in range(len(n_points)):
    axs[0][i].plot(S1_bettis[i].mean(axis=0))
    axs[0][i].set(xlabel=f"n_points: {n_points[i]}", ylabel='mean')
for i in range(len(n_points)):
    axs[1][i].plot(S1_bettis[i].std(axis=0))
    axs[1][i].set(xlabel=f"n_points: {n_points[i]}", ylabel='std')
plt.suptitle('S^1 first betti')
plt.tight_layout()
plt.show()
```

```python
n_points = [4,8,16,32, 64, 128, 256, 512, 1024]
torus_bettis = [np.load(f'/mount/paiktj/first_betti_torus_{i}.npy') for i in n_points]
```
