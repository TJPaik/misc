```python
import json

import networkx as nx
import matplotlib.pyplot as plt

from IteratedCriticalComplexes.SimplicialComplex import Torus, DunceHat
from IteratedCriticalComplexes.functions import simp_tree_to_hasse_diag, find_matching, make_critical_complex
```

```python
def plot_graph(G):
    pos = nx.circular_layout(G)
    nx.draw_networkx(G, pos)
```

# 1. Torus example

```python
initial_simplicial_complex = Torus()

initial_hasse_diagram = simp_tree_to_hasse_diag(initial_simplicial_complex)
new_hasse_diagram = initial_hasse_diagram.copy()
```

```python
fig = plt.figure(figsize=(15, 15))
plot_graph(initial_hasse_diagram)
fig.suptitle('torus - original')
plt.show()
```

```python
Step = 0
while True:
    Step += 1
    print(Step)
    hasse_diagram = new_hasse_diagram
    hasse_diagram = find_matching(hasse_diagram)
    new_hasse_diagram = make_critical_complex(hasse_diagram)
    if len(new_hasse_diagram.edges) == 0:
        break
```

```python
fig = plt.figure(figsize=(15, 15))
plot_graph(new_hasse_diagram)
fig.suptitle('torus - after reduction')
plt.show()
```

```python
homology = dict()
for node in new_hasse_diagram.nodes:
    node = json.loads(node)
    if len(node) in homology:
        homology[len(node)] += 1
    else:
        homology[len(node)] = 1

print('Torus - betti numbers')
for dim, betti in sorted(homology.items(), key=lambda x: x[0]):
    print(f'H_{dim - 1} : {betti}', end='\t\t')
```
